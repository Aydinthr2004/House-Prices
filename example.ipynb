```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('./train.csv')
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.describe(include='object')
```

```python
for col in df.columns:
    if df[col].dtype != 'object':
        mean_value=df[col].mean()
        df.fillna(mean_value , inplace=True)
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder


label = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        # drop NaN and fit label
        df[col] = label.fit_transform(df[col].dropna().astype(str).values)
        
df.head()
```

```python
target=df['SalePrice']
df = df.drop('SalePrice' , axis=1)
df = df.drop('Id' , axis=1)
df.head()
```

```python
X=df.values
X.shape
```

```python
y=target
y
```

```python
from sklearn.preprocessing import StandardScaler ,MinMaxScaler , QuantileTransformer

scaler=QuantileTransformer(output_distribution='normal', random_state=42)
X=scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X ,y ,random_state=42 , test_size=0.2)
```

# Random Forrest

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(random_state=42 , max_depth=16 , min_samples_split=2 ,n_estimators=200)
rf.fit(X_train , y_train)
```

```python
from sklearn.metrics import r2_score

y_pred_train_rf=rf.predict(X_train)
y_pred_test_rf=rf.predict(X_test)

r2_score(y_true=y_test , y_pred=y_pred_test_rf)
```

```python
r2_score(y_true=y_train , y_pred=y_pred_train_rf)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dt=DecisionTreeRegressor(random_state=42 , max_depth=32)
dt.fit(X_train , y_train)
```

```python
y_pred_train_dt=dt.predict(X_train)
y_pred_test_dt=dt.predict(X_test)

r2_score(y_true=y_test , y_pred=y_pred_test_dt)
```

```python
r2_score(y_true=y_train , y_pred=y_pred_train_dt)
```

# Lasso

```python
from sklearn.linear_model import Lasso

lasso=Lasso(alpha=0.0001, max_iter=10000, tol=1e-4, random_state=42)
lasso.fit(X_train,y_train)
```

```python
y_pred_train_lasso=lasso.predict(X_train)
y_pred_test_lasso=lasso.predict(X_test)

r2_score(y_true=y_test , y_pred=y_pred_test_lasso)
```

```python
r2_score(y_true=y_train , y_pred=y_pred_train_lasso)
```

# Grid search

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

gbr=GradientBoostingRegressor()

model = {
    'n_estimators' : np.arange(800 , 1200 , 200),
    'learning_rate' : np.arange(0.01 , 0.11 , 0.05)
}

grid=GridSearchCV(estimator=gbr, param_grid=model , cv=2)
grid.fit(X_train,y_train)
grid.best_estimator_
```

```python
gbr=GradientBoostingRegressor(learning_rate= 0.060000000000000005 , n_estimators=800)
gbr.fit(X_train,y_train)
```

```python
y_pred_train_gbr=gbr.predict(X_train)
y_pred_test_gbr=gbr.predict(X_test)

r2_score(y_true=y_test , y_pred=y_pred_test_gbr)
```

```python
r2_score(y_true=y_train , y_pred=y_pred_train_gbr)
```
